## 1.4 - Basic Text Processing Techniques

In this section we explore key text processing techniques essential for nlp tasks. These techniques help clean and prepare text data, making it easier to analyze

### 1. Tokenization

Tokenization involves splitting text into smaller units like words or sentenses. This is often the first step in text processing

```python
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt') # Ensure the tokenizer is available

# Example text
text = 'NLP is fun and useful!'

# Tokenize the text
tokens = word_tokenize(text)
print("Tokens: ", tokens)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords') # Ensure stopwords are available

# Defining a set of stop words 
stop_words = set(stopwords.words('english'))

# Filter the stopwords from the tokenized text
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
print("Filtered Tokens: ", filtered_tokens)
```

```python
text = "The quick brown fox jumps over the lazy dog"
text_tokens = word_tokenize(text)
filtered_text_tokens = [word for word in text_tokens if word.lower() not in stop_words]
print("Filtered Text Tokens: ", filtered_text_tokens)
```

```python
new_text = "Once upon a time there lived a king. This king has 7 sons. One day all of them went to hunt and hunted 7 fishes"
new_tokens = word_tokenize(new_text)
new_filtered_tokens = [word for word in new_tokens if word.lower() not in stop_words]
print("7 Fishes story filtered: ", new_filtered_tokens)
```

### Stemming 
Stemming reduces words to their root form by chopping off suffixes. It helps in grouping words with the same meaning

```python
from nltk.stem import PorterStemmer

# Initialize the stemmer
stemmer = PorterStemmer()

# Stem the filtered tokens
stemmed_tokens = [stemmer.stem(token) for token in filtered_tokens]
print("Stemmed Tokens: ", stemmed_tokens)
```
