## 1.5 - Hands-on Project: Text Preprocessing Pipeline

In this mini project, we will build a text preprocessing pipeline from scratch. The pipeline will take raw text as input and perform several steps of preprocessing to clean and prepare it for analysis. 

The Goal is to apply all the concepts we have learned so far: tokenization, stopword removal, stemming, lemmatization and case conversion.

We will also provide options for switching between stemming and lemmatization

### 1. Problem Statement

When working with text data (such as product reviews, tweets, or news articles), the text is often noisy and contains unnecessary information.

Preprocessing the text helps clean it and makes it suitable for tasks like sentimental analysis, classification or topic modeling.

#### Objective:

Develop a text preprocessing pipeline that:
   1. Tokenize the input text
   2. Remove Stopwords
   3. Convert text to lowercase
   4. Applies either stemming or lemmatization
   5. Returns the cleaned text ready for analysis

### 2. Step-by-step Plan
We will follow these steps :
   1. Input: Take raw text as input from the user.
   2. Tokenization: Split the raw input into tokens.
   3. Stopword Removal: Filter out common stopwords.
   4. Case Conversion: Convert all text to lowercase.
   5. Stemming/Lemmatization: Reduce words to their root form(based on user choise)

```python
# import required libraries
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords

# Download the necessary resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

### 3. Implementation

We will now define the preprocessing pipeline. The pipeline will take a peice of text as input and allow the user to choose between Stemming 
and lemmatization.

```python
# Define the preprocessing pipeline.
def preprocess_text(text, use_lemmatization=True):
    # Tokenization
    tokens = word_tokenize(text)

    # Stopword removal
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    # Case Conversion
    tokens = [token.lower() for token in tokens]

    # Stemming or Lemmatization
    if use_lemmatization:
        lemmatizer = WordNetLemmatizer()
        tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]

    else:
        stemmer = PorterStemmer()
        tokens = [stemmer.stem(word) for word in tokens if word not in stop_words]

    return tokens
```

### 4. Test the Pipeline

Lets test our pipeline on a sample input

```python
text = "A quick brown fox jumps over a lazy dog"
print("With Lemmatization: ",preprocess_text(text, use_lemmatization=True))
print("With Stemming: ",preprocess_text(text, use_lemmatization=False))
```

```python
test_text = """
Beast machine I brought the 16" rtx 5070 intel core ultra 9 285h 32 gb ram variant. Its a beast of a laptop. It can handle any game you throw at it. Most games i played gave above 100 fps. All AAA title games i played at ultra graphic settings. The thermals of this laptop is top notch. At ultra graphic settings also it's noise levels are very low. And during normal usage it's very silent. The pre installed omen gaming hub is very useful for changing performance modes during gaming. At this specs aus acer lenovo are charging 1.9 lakhs, but due to corporate login and sale offer at hp it was available at 1.4 lakhs, which is a steal deal. Literally best laptop at this price range. You can check it's reviews on youtube too.
"""
res = preprocess_text(test_text, use_lemmatization=True)
print(res)
print(len(res))
```
