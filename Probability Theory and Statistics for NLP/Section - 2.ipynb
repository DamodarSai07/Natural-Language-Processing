## 2.1 - Introduction to Probabiltiy for NLP

Probability theory forms the backbone of many Natural Language Processing (NLP) tasks. Understanding probability allows us to model 

uncertainities and predict the likelihood of events occuring in NLP, Probability helps us

1. Predict Words: Given a sequence of words, what is the probability of next word? This is crucial in tasks like text generation, auto-
   completion and language modelling.
2. Measure Word Association: Probability can help us determine how likely certain words are to appear together, which helps in semantic analysis.
3. Evaluate Language Models: Probability scores provide the way to access the performance of different models based on how well they predict real-world language usage

This section provides an introduction to some functional probability concepts that are directly relavent to NLP.

### Key Terms

Before diving into probability concepts, a few key terms used inside NLP

  1. Corpus: A corpus is a large collection of text used for language analysis and training NLP models. A corpus serves as a dataset that represents        the language that we want our model to understand. It can be a collection of books, articles, tweets, the British National Corpus (BNC) and            Google Ngrams are well-known corpura.
  2. Token: A token is a single unit of text, such as word or punctuation mark. In NLP, text data is often broken down into tokens for analysis and         procesing. For instance the sentence "NLP is Fun!" can be broken into ['NLP', 'is', 'Fun', '!']
  3. N-Grams: An n-gram is a contiguous sentence of 'n' items (typically words or characters) from a given sample of text. For example, in the phrase       "the cat sat", the 2-grams (bigrams) would be ["the cat", "cat sat"].

These terms are foundational as we discuss probability in the context of NLP

### Basic Probability

The basic probability of an event is defined as the ratio number of favourite outcomes to total number of possible outcomes.

Probability values ranges between 0(impossible event) and 1(certain event). This concept can be applied to any situation where we need to quantify the likelihood of something hapenning 

## Formula:                    
                        
              P(Event) = (Number of favourable outcomes)/Total Number of possible outcomes


```python
# Coin toss example
import random
from collections import Counter

# Example tossing a coin
def coin_toss():
    return random.choice(["Heads", "Tails"])

# Simulating 100 coin tosses 
results = [coin_toss() for _ in range(100)]
prob_heads = results.count("Heads")/len(results)
prob_tails = results.count("Tails")/len(results)
print(f"Probability of getting heads is {prob_heads}")
print(f"Probability of getting tails is {prob_tails}")
```

```python
def odd_even():
    return random.choice(["odd", "even"])

# Simulating 1000 odd or even
res = [odd_even() for _ in range(1000)]
prob_odd = res.count("odd")/len(res)
prob_eve = 1 - prob_odd
print(f"Probability of getting odd is {prob_odd}")
print(f"Probability of getting even is {prob_eve}")
```

```python
def roll_dice():
    return random.choice([1, 2, 3, 4, 5, 6])

# Simulating 100 dice rolls
resu = [roll_dice() for _ in range(100000)]
prob_one = resu.count(1)/len(resu)
prob_two = resu.count(2)/len(resu)
prob_three = resu.count(3)/len(resu)
prob_four = resu.count(4)/len(resu)
prob_five = resu.count(5)/len(resu)
prob_six = resu.count(6)/len(resu)

print(f"Probability of getting 1 is {prob_one}")
print(f"Probability of getting 2 is {prob_two}")
print(f"Probability of getting 3 is {prob_three}")
print(f"Probability of getting 4 is {prob_four}")
print(f"Probability of getting 5 is {prob_five}")
print(f"Probability of getting 6 is {prob_six}")
```

```python
def will_rain():
    return random.choice(["Yes", "No"])

# consider this probability for 10 years = 3650

ans = [will_rain() for _ in range(3650)]
prob_yes = ans.count("Yes")/len(ans)
prob_no = ans.count("No")/len(ans)
print(f"Probability of getting rain is {prob_yes}")
print(f"Probability of not getting rain is {prob_no}")
```

```python
# consider team a has 40% chances of winning and team b has 50% chances of winning and another team c has only 10% chances of winning 
# now find probability of winning if three teams plays 100 matches each 
def will_win():
    return random.choice([0.4, 0.5, 0.1])

# consider 100 matches
answer = [will_win() for _ in range(100)]
a_win = answer.count(0.4)/len(answer)
b_win = answer.count(0.5)/len(answer)
c_win = answer.count(0.1)/len(answer)
print(f"Probability of winning for A: {a_win}")
print(f"Probability of winning for B: {b_win}")
print(f"Probability of winning for C: {c_win}")
```
